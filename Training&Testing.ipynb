```python
import pandas as pd
```

```python
dataset = pd.read_csv('Standardized.csv')
dataset
```

```python
dataset.dtypes
```

```python
shuffledDataset = dataset.sample(frac=1).reset_index(drop=True)
shuffledDataset = shuffledDataset.sample(frac=1).reset_index(drop=True)
shuffledDataset = shuffledDataset.sample(frac=1).reset_index(drop=True)
shuffledDataset = shuffledDataset.sample(frac=1).reset_index(drop=True)
shuffledDataset = shuffledDataset.sample(frac=1).reset_index(drop=True)
shuffledDataset
```

```python
import numpy as np
```

```python
rename = {'Flow Duration': 'Flow Duration', 'TotLen Fwd Pkts': 'Source Bytes', 'TotLen Bwd Pkts': 'Destination Bytes', 'Tot Fwd Pkts': 'Source Packets', 'Tot Bwd Pkts': 'Destination Packets', 'Flow Byts/s': 'Flow Bytes/s', 'Flow Pkts/s': 'Flow Packets/s'}

shuffledDataset.rename(columns=rename, inplace=True)

dataset1, dataset2, dataset3, dataset4, dataset5 = np.array_split(shuffledDataset, 5)

predictors = ['Flow Duration', 'Source Bytes', 'Destination Bytes', 'Source Packets', 'Destination Packets', 'Flow Bytes/s', 'Flow Packets/s', 'IPv6', 'TCP', 'UDP']

shuffledDataset
```

```python
from sklearn.metrics import accuracy_score as AS, precision_score as PS, recall_score as RS, f1_score as F1S
from sklearn.metrics import classification_report
```

RANDOM FOREST CLASSIFICATION

```python
from sklearn.ensemble import RandomForestClassifier

RandomForestTrainSet = pd.concat([dataset2, dataset3, dataset4, dataset5])
RandomForestTestSet = dataset1

rfc = RandomForestClassifier()

rfc.fit(RandomForestTrainSet[predictors], RandomForestTrainSet['Label'])
RandomForestPredictions = rfc.predict(RandomForestTestSet[predictors])

rfcAS1 = AS(RandomForestTestSet['Label'], RandomForestPredictions)
rfcPS1 = PS(RandomForestTestSet['Label'], RandomForestPredictions, average='weighted')
rfcRS1 = RS(RandomForestTestSet['Label'], RandomForestPredictions, average='weighted')
rfcF1S1 = F1S(RandomForestTestSet['Label'], RandomForestPredictions, average='weighted')

RandomForestOutput = pd.concat([RandomForestTestSet['Label'], pd.Series(RandomForestPredictions, index=RandomForestTestSet.index)], axis=1)
RandomForestOutput.columns = ['Actual', 'Predictions']
RandomForestOutput
```

```python
print (classification_report(RandomForestOutput['Actual'], RandomForestOutput['Predictions']))
```
